```python
file_b = 'TD_POP1B_2021.xlsx'

pd.set_option('display.max_columns', None)
E = pd.read_excel(file_b, sheet_name="COM",skiprows = 10)

print(len(E))
E[:2]
```

```python
E.columns = [col.replace('SEXE1_AGED100', 'SEXE1_AGED_') if col.startswith('SEXE1_AGED100') else col for col in E.columns]
E.columns = [col.replace('SEXE2_AGED100', 'SEXE2_AGED_') if col.startswith('SEXE2_AGED100') else col for col in E.columns]
print(len(E))
E[:2]
```

```python
columns_to_sum = E.filter(like='SEXE1_AGED_')
sum_column = columns_to_sum.sum(axis=1)
E = pd.concat([E, sum_column.rename('Nb_H')], axis=1)

columns_to_sum = E.filter(like='SEXE2_AGED_')
sum_column = columns_to_sum.sum(axis=1)
E = pd.concat([E, sum_column.rename('Nb_F')], axis=1)
print(len(E))
E[:2]
```

```python
column_to_move = 'Nb_H'
columns = E.columns.tolist()
columns.insert(2, columns.pop(columns.index(column_to_move)))
E =E[columns]

column_to_move = 'Nb_F'
columns = E.columns.tolist()
columns.insert(3, columns.pop(columns.index(column_to_move)))
E =E[columns]
print(len(E))
E[:2]
```

### Selection des 25 villes les plus peuplées

```python
F['Total'] = F['Nb_H'] + F['Nb_F']
top_25_cities = F.nlargest(25, 'Total').drop(columns=['Total'])
print (len(top_25_cities))
top_25_cities[:2]
```

```python
# Select population columns for males (SEXE1) and females (SEXE2)
population_columns_H = [col for col in top_25_cities.columns if col.startswith('SEXE1_AGED_')]
population_columns_F = [col for col in top_25_cities.columns if col.startswith('SEXE2_AGED_')]

# Common columns to keep during melting
id_vars = ['CODGEO', 'LIBGEO',  'code_dept', 'nom_departement', 'code_reg', 'nom_region']

# Melt DataFrames for males and females
top_25_cities_melted_H = top_25_cities.melt(id_vars=id_vars, value_vars=population_columns_H, var_name='Age', value_name='value_H')
top_25_cities_melted_F = top_25_cities.melt(id_vars=id_vars, value_vars=population_columns_F, var_name='Age', value_name='value_F')

# Extract Age as an integer
top_25_cities_melted_H['Age'] = top_25_cities_melted_H['Age'].str[-3:].astype(int)
top_25_cities_melted_F['Age'] = top_25_cities_melted_F['Age'].str[-3:].astype(int)

# Rename columns
top_25_cities_melted_H.rename(columns={'value_H': 'Nb_H'}, inplace=True)
top_25_cities_melted_F.rename(columns={'value_F': 'Nb_F'}, inplace=True)

# Merge on common columns (excluding the different population counts)
merge_columns = ['CODGEO', 'LIBGEO', 'Age', 'code_dept', 'nom_departement', 'code_reg', 'nom_region']
top_25_cities_melted = pd.merge(top_25_cities_melted_H, top_25_cities_melted_F, on=merge_columns, how='inner')

# Convert relevant columns to integers
top_25_cities_melted[['Nb_H', 'Nb_F']] = top_25_cities_melted[['Nb_H', 'Nb_F']].astype(int)

#del top_25_cities_melted[top_25_cities_melted.columns[0]]
# Display result
print(top_25_cities_melted)
```
